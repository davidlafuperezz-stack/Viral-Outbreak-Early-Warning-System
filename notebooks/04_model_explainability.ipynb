# 04 - Model Explainability and Interpretation

This notebook documents the explainability stage of the **Viral Outbreak Early Warning System** project.

The objective is to understand which variables contribute most strongly to 7-day outbreak risk prediction.

Explainability is especially important in epidemiological surveillance because predictive systems should not only produce risk scores, but also provide interpretable evidence that can support public health decision-making.

## 1. Why Explainability Matters

In healthcare and epidemiology, model interpretability is critical.

A model that predicts outbreak risk should help answer questions such as:

- Which variables are associated with increased outbreak risk?
- Are predictions driven by epidemiological signals, environmental conditions or mobility patterns?
- Does the model rely on plausible public health indicators?
- Can results be explained to non-technical stakeholders?

For this reason, this project includes both model-agnostic and model-specific interpretability methods.

## 2. Explainability Methods Used

This project uses three complementary approaches:

### Permutation Importance

Measures how much model performance decreases when a feature is randomly shuffled.

### Model-Based Feature Importance

Available for some tree-based models, where feature importance is estimated from the fitted model.

### SHAP Values

SHAP estimates how each feature contributes to individual model predictions.

Together, these methods provide a more complete view of model behavior.

## 4. Load Permutation Importance Results

Permutation importance is useful because it evaluates how much each feature contributes to model performance on unseen data.

````python
PERMUTATION_PATH = Path("reports/tables/permutation_importance.csv")

permutation_importance = pd.read_csv(PERMUTATION_PATH)
permutation_importance.head(15)
````

## 5. Top Predictive Features

The following table shows the most influential features according to permutation importance.

````python
top_permutation = permutation_importance.head(15)

top_permutation
````

````python
plt.figure(figsize=(9, 6))

plot_data = top_permutation.sort_values("importance_mean")

plt.barh(
    plot_data["feature"],
    plot_data["importance_mean"],
    xerr=plot_data["importance_std"]
)

plt.title("Top Features by Permutation Importance")
plt.xlabel("Decrease in Precision-Recall AUC")
plt.ylabel("Feature")

plt.tight_layout()
plt.show()
````

## 9. Interpreting the Results

Permutation importance helps identify which variables are most relevant for prediction.

In a realistic outbreak early warning system, we would expect important variables to include:

- recent case counts
- incidence per 100,000
- effective reproduction rate
- spatial pressure
- mobility
- healthcare pressure
- lagged and rolling case indicators

If these variables appear among the most important predictors, it supports the plausibility of the simulation and modeling pipeline.

## 10. Explainability Limitations

The explainability analysis should be interpreted carefully.

Important limitations include:

- The dataset is synthetic and does not represent a real outbreak.
- Feature importance does not prove causality.
- Correlated predictors may share importance.
- SHAP values depend on the trained model and feature preprocessing.
- Public health interpretation would require validation with real-world epidemiological data.

These limitations are important to report transparently.

## 11. Practical Interpretation for Public Health

In a real surveillance setting, explainability outputs could help decision-makers understand why a region is classified as high risk.

For example, a high-risk prediction may be driven by:

- increasing recent cases
- elevated spatial pressure from neighboring regions
- low vaccination rate
- high mobility
- high healthcare pressure
- unfavorable environmental conditions

This makes the model more useful as a decision-support tool rather than a black-box classifier.
